```python
import json

#lets write a function to load the data
def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
```

```python
data=load_data("data.json")
```

```python
def display_users(data):
    print("User and their freinds\n")
    for user in data["users"]:
        print(f"{user["name"]} is friends with:{user["friends"]}")
        
    print("User and their liked pages are\n")
    for user in data["users"]:
        print(f"{user["name"]} liked pages are:{user["liked_pages"]}")
    print("Page information\n")    
    for page in data["pages"]:
        print(f"{page["id"]}:{page["name"]}")
```

```python

display_users(data)
```
