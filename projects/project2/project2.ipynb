```python
def parse_chunk(chunk):
    chunk=chunk.strip()
    sep_chunk=chunk.split("\n")
    username=sep_chunk[0]
    no_of_post=sep_chunk[1]
    followers=sep_chunk[2]
    following=sep_chunk[3]
    name=sep_chunk[4]
    type_of_page=sep_chunk[5]
    bio="\n".join(sep_chunk[6:])
    print(username, no_of_post, followers, following, name, type_of_page, bio,sep="\n")

parse_chunk(chunks[0])
```
